## Import Libraries

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading the image which is in Array form.
img = cv.imread('kesarkar.jpg')
```

```python
plt.figure(figsize = (22,6))
plt.title('Original Image', size = 22)
plt.imshow(img[:,:,::-1])
plt.axis('off')
plt.show()
```

```python
# Converting image from BGR to RGB color
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize = (22,6))
plt.title('Original Image', size = 22)
img_grey = cv.cvtColor(img_grey, cv.COLOR_BGR2RGB)
plt.imshow(img_grey[:,:-1])
plt.axis('off')
plt.show()
```

```python
# Inverting the Image
img_invert = cv.bitwise_not(img_grey)
```

```python
plt.figure(figsize = (22,6))
plt.title('Original Image', size = 22)
img_invert = cv.cvtColor(img_invert, cv.COLOR_BGR2RGB)
plt.imshow(img_invert[:,:-1])
plt.axis('off')
plt.show()
```

```python
# Blurring the image using GaussianBlur function
img_blur = cv.GaussianBlur(img_invert, (111,111), 0)
```

```python
plt.figure(figsize = (22,6))
plt.title('Original Image', size = 22)
img_blur = cv.cvtColor(img_blur, cv.COLOR_BGR2RGB)
plt.imshow(img_blur[:,:-1])
plt.axis('off')
plt.show()
```

```python
# Inverting the blurred image
img_blur_invert = cv.bitwise_not(img_blur)
```

```python
# Converting the blurred inverted image to sketch by doing bitwise division
img_sketch = cv.divide(img_grey, img_blur_invert, scale = 256.0)
```

```python
# Plotting the Images
plt.figure(figsize = (14,8))
plt.title('Pencil Sketch', size = 18)
sketch_rgb = cv.cvtColor(img_sketch, cv.COLOR_BGR2RGB)
plt.imshow(sketch_rgb)
plt.axis('off')
plt.show()
```
